```python
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.by import By
import time


# 세팅
driver = webdriver.Chrome()
driver.set_window_size(800, 1000)

# 유튜브 열기
Url = 'https://www.youtube.com/watch?v=LKJwO9PRuBg'
driver.get(Url)
time.sleep(5)

# 유튜브 댓글 끝까지 로딩
last_page_height = driver.execute_script("return document.documentElement.scrollHeight")
while True:
    driver.execute_script("window.scrollTo(0, document.documentElement.scrollHeight);")
    time.sleep(1.0)
    new_page_height = driver.execute_script("return document.documentElement.scrollHeight")
    if new_page_height == last_page_height:
        break
    last_page_height = new_page_height
    
    # 유튜브 대댓글 열기
    elements = driver.find_elements(By.CSS_SELECTOR,"#more-replies")
    for element in elements :
        driver.execute_script("arguments[0].click();", element)
driver.execute_script("window.scrollTo(0, 0);")


# 파싱
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')

id_list = []
comment_list = []

comments = soup.find_all("ytd-comment-thread-renderer", class_ = "style-scope ytd-item-section-renderer")

for comment in comments :
    comment_text = comment.find("yt-attributed-string", id="content-text").text
    id = comment.find("span","style-scope ytd-comment-view-model style-scope ytd-comment-view-model").text
    id_list.append(id)
    comment_list.append(comment_text)

id_final = []
comment_final = []

for i in range(len(comments)):
    temp_id = id_list[i]
    temp_id = temp_id.replace('\n','')
    temp_id = temp_id.replace('\t','')
    temp_id = temp_id.replace('            ','')
    id_final.append(temp_id)
    temp_comment = comment_list[i]
    temp_comment = temp_comment.replace('\n', '')
    temp_comment = temp_comment.replace('\t', '')
    temp_comment = temp_comment.replace('    ', '')
    comment_final.append(temp_comment)


data = {"id" :id_final , "comment":comment_final }

print(data)
```
