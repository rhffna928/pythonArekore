```python
import pandas as pd
df_m = pd.read_excel('201401_201401_연령별인구현황_월간.xlsx', skiprows=3, index_col='행정기관', usecols='B,E:O')
df_m.head(5)
```

```python
df_m.iloc[0] = df_m.iloc[0].str.replace(',','').astype(int)
df_m
```

```python
df_m.iloc[0]
```

```python
df_w = pd.read_excel('201401_201401_연령별인구현황_월간.xlsx', skiprows=3, index_col='행정기관', usecols='B,R:AB')
df_w.head(5)
```

```python
df_w.iloc[0] = df_w.iloc[0].str.replace(',','').astype(int)
df_w
```

```python
df_w.columns = df_m.columns
df_w
```

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
plt.figure(figsize=(10,7))
plt.barh(df_m.columns, -df_m.iloc[0] //1000) # 단위 천명
plt.barh(df_w.columns, df_w.iloc[0] //1000) 
plt.title('2014 대한민국 인구 피라미드')
plt.savefig('2014 대한민국 인구 피라미드.png',dpi=200)
plt.show()
```
