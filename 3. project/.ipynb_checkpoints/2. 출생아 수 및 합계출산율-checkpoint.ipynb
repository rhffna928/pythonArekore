# 출생아 수 및 합계출산율

```python
import pandas as pd
df = pd.read_excel('bear avg.xlsx',skiprows=2, nrows=2, index_col=0)
df.index.values
```

```python
df.rename(index={'출생아\xa0수':'출생아 수','합계\xa0출산율':'합계 출산율'})
```

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
df.columns.values
```

```python
df.rename(columns={'출생아\xa0수':'출생아 수','합계\xa0출산율':'합계 출산율'}, inplace=True)
df
```

```python
df
```

```python

plt.bar(df.index,df['출생아 수'])
plt.plot(df.index,df['합계 출산율'])
```

```python
fig, ax1 = plt.subplots(figsize=(8,5))
ax1.plot(df.index,df['출생아 수'], c='#ff812d')
ax2 = ax1.twinx() # x축을 공유하는 쌍둥이 axis
ax2.plot(df.index,df['합계 출산율'], c='#ffd100')
```

```python
fig, ax1 = plt.subplots(figsize=(8,5))
ax1.set_ylabel('출생아 수 (천 명)')
ax1.set_ylim(250,700)
ax1.set_yticks([300,400,500,600])
ax1.bar(df.index,df['출생아 수'], color='#ff8120')
ax2 = ax1.twinx() # x축을 공유하는 쌍둥이 axis
ax2.set_ylabel('합계 출산율(가임여성 1명 당 명)')
ax2.set_ylim(0,2)
ax2.set_yticks([0,1])
ax2.plot(df.index,df['합계 출산율'], color='#ffd100')
```

```python
fig, ax1 = plt.subplots(figsize=(8,4))
ax1.set_ylabel('출생아 수 (천 명)')
ax1.set_ylim(250,700)
ax1.set_yticks([300,400,500,600])
ax1.bar(df.index,df['출생아 수'], color='#ff8120')
for idx, val in enumerate(df['출생아 수']):
    ax1.text(idx, val + 12, val, ha='center')

ax2 = ax1.twinx() # x축을 공유하는 쌍둥이 axis
ax2.set_ylabel('합계 출산율(가임여성 1명 당 명)')
ax2.set_ylim(0,2)
ax2.set_yticks([0,1])
ax2.plot(df.index,df['합계 출산율'], color='#ffd100', marker='o', ms=15,lw=3,mec='w',mew=3)
```

```python
fig, ax1 = plt.subplots(figsize=(8,4))
fig.suptitle('출생아 수 및 합계출산율')
ax1.set_ylabel('출생아 수 (천 명)')
ax1.set_ylim(250,700)
ax1.set_yticks([300,400,500,600])
ax1.bar(df.index,df['출생아 수'], color='#ff8120')
for idx, val in enumerate(df['출생아 수']):
    ax1.text(idx, val + 12, val, ha='center')

ax2 = ax1.twinx() # x축을 공유하는 쌍둥이 axis
ax2.set_ylabel('합계 출산율(가임여성 1명 당 명)')
ax2.set_ylim(0,2)
ax2.set_yticks([0,1])
ax2.plot(df.index,df['합계 출산율'], color='#ffd100', marker='o', ms=15,lw=3,mec='w',mew=3)
for idx, val in enumerate(df['합계 출산율']):
    ax2.text(idx, val + 0.15,val,ha='center')
```
