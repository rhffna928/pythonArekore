```python
import json
file1_path = 'path_trc_01.log'
file2_path = 'path_trc_T01.log'
output_file_path = 'output_file.json'

# 파일 1과 파일 2에서 모든 데이터를 읽어들임
file1_data = []
file2_data = []

with open(file1_path, 'r', encoding='utf-8') as file1:
    for line in file1:
        try:
            data = json.loads(line)
            file1_data.append(data)
        except json.JSONDecodeError:
            print(f"Error decoding JSON in file1: {line}")

with open(file2_path, 'r', encoding='utf-8') as file2:
    for line in file2:
        try:
            data = json.loads(line)
            file2_data.append(data)
        except json.JSONDecodeError:
            print(f"Error decoding JSON in file2: {line}")

# 교환된 데이터를 담을 리스트
output_data = []

# 두 파일에서 'TRACTOR_INFO' 데이터 교환
for data1, data2 in zip(file1_data, file2_data):
    if data1.get('type') == 'TRACTOR_INFO' and data2.get('type') == 'TRACTOR_INFO':
        # 'data' 부분만 교환
        data1['data'], data2['data'] = data2['data'], data1['data']
    
    # 교환된 데이터를 output_data에 추가
    output_data.append(data1)
    output_data.append(data2)

# 파일 1과 파일 2에서 나머지 데이터는 그대로 추가
# 이미 zip으로 교환된 'TRACTOR_INFO' 데이터는 제외하고, 나머지는 그대로 추가
output_data.extend(file1_data[len(file2_data):])  # file1에 file2보다 더 많은 데이터가 있을 경우
output_data.extend(file2_data[len(file1_data):])  # file2에 file1보다 더 많은 데이터가 있을 경우

# 결과를 새로운 파일에 저장
with open(output_file_path, 'w', encoding='utf-8') as output_file:
    for data in output_data:
        json.dump(data, output_file, ensure_ascii=False)
        output_file.write("\n")  # 각 JSON 객체를 한 줄씩 저장
```
