```python
import json

def interpolate_coordinates(lat1, lon1, lat2, lon2, N):
    lat_values = [lat1 + (lat2 - lat1) * i / (N + 1) for i in range(1, N + 1)]
    lon_values = [lon1 + (lon2 - lon1) * i / (N + 1) for i in range(1, N + 1)]
    return list(zip(lat_values, lon_values))

# Example usage

lat1, lon1 = 34.0522, -118.2437 # Los Angeles

lat2, lon2 = 36.1699, -115.1398 # Las Vegas

N = 1000

positions = interpolate_coordinates(lat1, lon1, lat2, lon2, N)

print(positions)
```

```python
#새로운 파일 만들고 json 파일에서 읽은 command 값에 따라 구분해서 파일 만들기
import json
import os

file_path = "./TractorInfo_000005.log"
with open(file_path, 'r' ,encoding='utf-8') as f:
    lines = f.readlines()
    
os.makedirs("output",exist_ok=True)

for line in lines:
    data =json.loads(line)
    command = data.get('command')
    if command:
        output_file = os.path.join('output',f"{command}.json")
        with open(output_file, 'a', encoding='utf-8') as of:
            json.dump(data,of)
            
            of.write('\n')
```
