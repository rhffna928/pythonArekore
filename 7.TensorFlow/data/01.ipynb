```python
# 차량 데이터 분석 및 모델 성능 향상 전략
# 데이터 로드 및 기본 분석
# 먼저 데이터를 로드하고 기본적인 정보를 확인하겠습니다.

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer

# 데이터 로드
df = pd.read_csv('./_VEHICLE_ING_INFO_250529stsw.csv', encoding='cp949')

# 데이터 기본 정보 확인
print(df.info())
print(df.describe())
print(df.isnull().sum())

# 데이터 전처리 (Data Preprocessing)
# 결측치 처리
# 이상치 처리
# 데이터 타입 변환
# 결측치 처리
df.fillna(df.median(numeric_only=True), inplace=True)  # 수치형 변수는 중앙값으로
df.fillna(df.mode().iloc[0], inplace=True)  # 범주형 변수는 최빈값으로

# 이상치 처리 (IQR 기준)
def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

# 수치형 열에 대해 이상치 제거
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns
for col in numeric_cols:
    df = remove_outliers(df, col)

# 데이터 타입 변환 (예시: 날짜 형식이 있다면 datetime으로 변환)
# df['date_column'] = pd.to_datetime(df['date_column'])

# 피처 엔지니어링 (Feature Engineering)
# 새로운 특성 생성
# 범주형 변수 인코딩
# 특성 선택
# 새로운 특성 생성 예시 (실제 데이터에 맞게 조정 필요)
# 예: 차량 연식과 현재 연도의 차이
# df['vehicle_age'] = pd.datetime.now().year - df['model_year']

# 범주형 변수 인코딩
categorical_cols = df.select_dtypes(include=['object']).columns
df_encoded = pd.get_dummies(df, columns=categorical_cols, drop_first=True)

# 특성 중요도 분석을 통한 특성 선택
X = df_encoded.drop('target_column', axis=1)  # 실제 타겟 변수명으로 변경 필요
y = df_encoded['target_column']

model = RandomForestRegressor()
model.fit(X, y)
feature_importances = pd.Series(model.feature_importances_, index=X.columns)
feature_importances.nlargest(10).plot(kind='barh')
plt.title('Top 10 Important Features')
plt.show()

# 중요도가 낮은 특성 제거
selected_features = feature_importances[feature_importances > 0.01].index
X_selected = X[selected_features]

# 하이퍼파라미터 튜닝 (Hyperparameter Tuning)
# 데이터 분할
X_train, X_test, y_train, y_test = train_test_split(X_selected, y, test_size=0.2, random_state=42)

# 파이프라인 구성
numeric_transformer = Pipeline(steps=[
    ('scaler', StandardScaler())])

categorical_transformer = Pipeline(steps=[
    ('onehot', OneHotEncoder(handle_unknown='ignore'))])

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_cols),
        ('cat', categorical_transformer, categorical_cols)])

# 모델 파이프라인
pipeline = Pipeline(steps=[('preprocessor', preprocessor),
                          ('model', RandomForestRegressor())])

# 하이퍼파라미터 그리드
param_grid = {
    'model__n_estimators': [100, 200, 300],
    'model__max_depth': [None, 10, 20, 30],
    'model__min_samples_split': [2, 5, 10],
    'model__min_samples_leaf': [1, 2, 4]
}

# 그리드 서치
grid_search = GridSearchCV(pipeline, param_grid, cv=5, scoring='neg_mean_squared_error', n_jobs=-1)
grid_search.fit(X_train, y_train)

# 최적 파라미터
print("Best parameters: ", grid_search.best_params_)

# 최적 모델 평가
best_model = grid_search.best_estimator_
y_pred = best_model.predict(X_test)
print("RMSE: ", np.sqrt(mean_squared_error(y_test, y_pred)))
print("R2 Score: ", r2_score(y_test, y_pred))

# 모델 성능 향상을 위한 추가 전략
# 교차 검증: K-Fold 교차 검증을 통해 모델의 일반화 성능 평가
# 앙상블 방법: 여러 모델을 결합하여 성능 향상 (예: Gradient Boosting, Stacking)
# 차원 축소: PCA 또는 t-SNE를 사용하여 고차원 데이터 처리
# 불균형 데이터 처리: SMOTE 또는 ADASYN을 사용한 오버샘플링
# 정규화: L1/L2 정규화를 적용하여 과적합 방지
# 교차 검증 예시
from sklearn.model_selection import cross_val_score
scores = cross_val_score(best_model, X_selected, y, cv=5, scoring='r2')
print("Cross-validated R2 scores: ", scores)
print("Mean R2: ", np.mean(scores))

# 앙상블 모델 예시 (XGBoost)
from xgboost import XGBRegressor
xgb_model = XGBRegressor()
xgb_model.fit(X_train, y_train)
y_pred_xgb = xgb_model.predict(X_test)
print("XGBoost R2 Score: ", r2_score(y_test, y_pred_xgb))

# 결론
# 데이터 전처리: 결측치와 이상치를 적절히 처리하여 데이터 품질 향상
# 피처 엔지니어링: 새로운 특성 생성과 중요도 기반 특성 선택으로 모델 입력 개선
# 하이퍼파라미터 튜닝: 그리드 서치를 통해 최적의 모델 파라미터 찾음
# 고급 기법 적용: 교차 검증, 앙상블 방법 등으로 추가 성능 향상 가능
# 이러한 단계를 통해 모델의 예측 정확도와 일반화 성능을 크게 향상시킬 수 있습니다. 실제 적용 시에는 비즈니스 목표와 데이터 특성에 맞게 각 단계를 조정해야 합니다.
```
